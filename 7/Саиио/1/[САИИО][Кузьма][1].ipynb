```python

def solve(A_inv, x, k):
  n = len(A_inv)
  k -= 1 # для лучшей индексации
  # ШАГ 1
  l = np.dot(A_inv, x)
  li = l[k] # находим i-ую компоненту вектора l
  if (li == 0): # проверка компоненты на равенство нолю
    return "Матрица не может быть обратимой"
  # ШАГ 2
  l1 = np.copy(l) # получаем вектор l с волной
  l1[k] = -1 # заменяем i-ую компоненту вектора l с волной -1
  # ШАГ 3
  l2 = np.dot(-1/li, l1) # получаем вектор l с шапочкой
  # ШАГ 4
  E = np.eye(n) # создаем единичную матрицу
  Q = np.copy(E)
  #print(l2)
  Q[:,k] = l2.transpose() # создаем матрицу Q
  z = np.eye(n)
  for i in range(n):
    for j in range(n):
      z[i][j] = Q[i][i]*A_inv[i][j]
      if (i != k):
        z[i][j] += Q[i][k]*A_inv[k][j]
  return z
```

```python
def get_matrix(av, Jb):
  a = np.eye(len(Jb))
  j = 0
  for i in Jb:
    a[:, j] = av[:, i - 1]
    j += 1
  a = np.linalg.inv(a)
  return a
```

```python
def main_simplex(A, c, x, jb):
  iteration = 0
  while True:
    iteration += 1
    if iteration == 1:
      ab = np.array([np.copy(A[:,i-1]) for i in jb]).transpose()
      A_i = np.linalg.inv(ab)
    else:
      ab[:,position_min_tetta] = A[:, jb[position_min_tetta] - 1]
      A_i = solve(A_i, ab[:,position_min_tetta], position_min_tetta + 1)
    cb = np.array([c[i - 1] for i in jb])
    u = np.dot(cb, A_i)
    delta = np.dot(u, A) -c
    j0=0
    if min(delta) >= 0:
      
      return (x, jb)
    while delta[j0]>=0:
      j0 += 1
    z = np.dot(A_i, A[:,j0])
    tetta = [float(x[jb[i] - 1])/float(z[i]) if z[i]> 0 else np.inf for i in range(len(jb))]
    tetta0 = np.min(tetta)
    if tetta0 is np.inf:
      print("целевая функция неограничена на множестве допустимых планов")
    position_min_tetta = np.where(tetta == tetta0)[0][0]
    jb_new = np.copy(jb)
    jb_new[position_min_tetta] = j0 + 1
    x_new = np.copy(x)
    j = 0
    x_new = x_new.astype(np.float64)
    for i in jb:
      x_new[i - 1] = float(x[i - 1]) - tetta0*float(z[j])
      j+=1
    x_new[j0] = tetta0
    x = x_new
    jb = jb_new
```

```python
import numpy as np

def read_specific_rows(path, rows=[]):
    with open(path) as f:
        for i, line in enumerate(f):
            if i in rows:
                yield line

def lw_1(matrix, index, vector):
    n = len(matrix)
    l_1 = matrix.dot(vector)
    if l_1[index - 1] == 0:
        return "NO"
    else:
        l_2 = l_1.copy()
        l_2[index - 1] = -1
        l_3 = (-1 / l_1[index - 1]) * l_2
        Q = np.identity(n)
        Q[:, index - 1] = l_3[:]
        #return Q.dot(matrix)
        res = np.identity(n)
        for i in range(n):    
            for j in range(n):
                if (i == index - 1):
                    res[i][j] = Q[i][i] * matrix[i][j]
                else:
                    res[i][j] = Q[i][i] * matrix[i][j] + Q[i][index - 1] * matrix[index - 1][j]
        return res

def get_θ(z, x, J_basis):
    θ = []
    for i in range(len(J_basis)):
        if z[i] > 0:
            θ.append(x[J_basis[i] - 1] / z[i])
        else:
            θ.append(np.inf)
    return θ

def lw_2(A, c, x, J_basis):
    is_first_iteration = True
    A_basis = np.zeros((A.shape[0], len(J_basis)))
    for i, J in enumerate(J_basis):
        A_basis[:, i] = A[:, J-1] 
    A_basis = A_basis.transpose()
    A_inv = np.linalg.inv(A_basis)
    while True:
        if not is_first_iteration:
            A_basis[:, θ_0_index] = A[:, J_basis[θ_0_index] - 1]
            A_inv = lw_1(A_inv, θ_0_index + 1, A_basis[:, θ_0_index])
        else:
            is_first_iteration = False
        c_basis = np.array([c[i - 1] for i in J_basis])
        u = c_basis.dot(A_inv)
        Δ = u.dot(A) - c
        if min(Δ) >= 0:
            return (x, J_basis)
        else:
            J_0 = 0
            for i in Δ:
                if i < 0:
                    break
                else:
                    J_0 += 1
            z = A_inv.dot(A[:, J_0])
            θ = get_θ(z, x, J_basis)
            θ_0 = np.min(θ)
            if θ_0 is np.inf:
                return "UNBOUNDED"
            θ_0_index = 0
            for i in θ:
                if i == θ_0:
                    break
                else:
                    θ_0_index += 1
            new_J_basis = np.copy(J_basis)
            new_J_basis[θ_0_index] = J_0 + 1
            new_x = np.copy(x)
            for i, J in enumerate(J_basis):
                new_x[J - 1] -= θ_0 * z[i]
            new_x[J_0] = θ_0
            x, J_basis = new_x, new_J_basis

#Случай 1
def case_1(new_A, new_A_inv, J_basis, k, n):
    for i in range(1, n + 1): 
        if i not in J_basis:
            if new_A_inv.dot(new_A[:, i - 1])[k] != 0:
                J_basis[k] = i
                return True
    return False

#Случай 2
def case_2(A, new_A, J_basis, k, n):
    i = J_basis[k] - n
    return np.delete(A, i, 0), np.delete(new_A, i, 0), np.delete(J_basis, k, 0)

def lw_3(A, b):
    n, m = A.shape[1], b.shape[0]
    for i, e in enumerate(b):
        if e < 0:
            b[i] *= -1
            A[i] *= -1
    new_A = np.hstack((A.copy(), np.eye(m)))   
    x = np.append(np.zeros(n), b)
    J_basis = np.zeros(m, dtype = int)
    for i in range(m):
        J_basis[i] = n + i + 1
    c = np.zeros(n + m, dtype = int)
    for i in range(n + m):
        if i < n:
            c[i] = 0
        else:
            c[i] = -1
    x, J_basis = lw_2(new_A, c, x, J_basis)
    for i in x[n:]:
        if i != 0:
            return 'INCONSISTENT'
    while True:
    	  J_basis_len = len(J_basis)
        k = None
      
        for i in range(J_basis_len):
            if (J_basis[i] > n):  
                k = i
                break
        if k == None:
            return x[:n]
        else:
            new_A_inv = np.eye(J_basis_len)
            for i, j in enumerate(J_basis):
                new_A_inv[:, i] = new_A[:, i - 1]
            new_A_inv = np.linalg.inv(new_A_inv)
            if not case_1(new_A, new_A_inv, J_basis, k, n):
                A, new_A, J_basis = case_2(A, new_A, J_basis, k, n)

A = np.array([[2,11],[1,1],[4,-5]])
b = np.array([38,7,5])
print("Input:")
print("A:\n", A)
print("b:\n", b)
print("Output:")
print(lw_3(A, b))
```
